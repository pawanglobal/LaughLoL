# Merge, Join, Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\PAWAN KALER\Downloads\03_LEARNINGS\LaughLoL\DA\DataAnalysisBootCamp\Python\Pandas\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\PAWAN KALER\Downloads\03_LEARNINGS\LaughLoL\DA\DataAnalysisBootCamp\Python\Pandas\LOTR 2.csv")
df2
```

```python
# Default inner
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how='inner', on = 'FellowshipID')
```

```python
df1.merge(df2, how='inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='right')
```

```python
df1.merge(df2, how='cross')
```

# Join function

```python
df1.join(df2, on='FellowshipID', how='outer', rsuffix='_right',lsuffix='_left')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), rsuffix='_right', lsuffix='_left', how='outer')
df4
```

# Concatenate

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], join='inner', axis=1)
```

```python
pd.concat([df1, df2], join='outer', axis=1)
```
