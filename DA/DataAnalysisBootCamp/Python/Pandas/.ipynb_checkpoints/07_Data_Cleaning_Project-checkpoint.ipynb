# Data Cleaning in Pandas
# We need to give the contact numbers of those people to a call center whom they can call

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\PAWAN KALER\Downloads\03_LEARNINGS\LaughLoL\DA\DataAnalysisBootCamp\Python\Pandas\Customer Call List.xlsx")
```

```python
df
```

```python
# Drop the duplicates
```

```python
df = df.drop_duplicates()
df
```

```python
# Droping columns which we don't want to use, we are not gonna clean them
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
# Column "Last_Name" has some irregularities; standardized it
```

```python
df["Last_Name"].str.strip()
```

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df
```

```python
# df["Last_Name"] = df["Last_Name"].str.lstrip("/")
# df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"] = df["Last_Name"].str.strip("_./")
df
```

```python
# Now, let's standardize the next "Phone_Number" column
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^A-Za-z0-9]', '', regex=True)
```

```python
df
```

```python
# df['Phone_Number'].dtype
```

```python
# Formatting the Phone_Number
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6]+ '-' + x[6:10])
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
df[['Street_Address','State','Zip_Code']]=df["Address"].str.split(',', expand=True)
df
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"]=df["Paying Customer"].str.replace('No','N')
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df=df.replace('N/a','')
df
```

```python
df=df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == "Y":
        df.drop(x, inplace=True)
df

#Another way to drop null values
#df = df.dropna(subset="Phone_Number"),inplace=Ture)
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == "":
        df.drop(x, inplace=True)
df     
```

```python
df = df.reset_index(drop=True)
df
```
